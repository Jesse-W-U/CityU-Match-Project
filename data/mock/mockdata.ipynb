```python
import pandas as pd
import random
import json
from datetime import datetime
```

```python
# student

print("üìä ÁîüÊàê student.csv (200‰∫∫)...")
students = []

COLLEGES = [
    'College of Business',
    'College of Liberal Arts and Social Sciences',
    'College of Science',
    'College of Engineering',
    'College of Veterinary Medicine and Life Sciences',
    'Other'
]
MAJORS = {
    'College of Business': ['Accounting', 'Finance', 'Marketing', 'Management'],
    'College of Engineering': ['Computer Science', 'Electronic Engineering', 'Mechanical Engineering', 'Biomedical Engineering'],
    'College of Science': ['Mathematics', 'Physics', 'Chemistry', 'Biology'],
    'College of Liberal Arts and Social Sciences': ['Media and Communication', 'Psychology', 'Economics', 'English'],
    'College of Veterinary Medicine and Life Sciences': ['Veterinary Medicine', 'Biomedical Sciences'],
    'Other': ['Interdisciplinary Studies']
}

# ‰∏≠ÂõΩÁúÅ‰ªΩÂàóË°®
PROVINCES = [
    "Beijing", "Shanghai", "Guangdong", "Zhejiang", "Jiangsu", "Sichuan",
    "Hubei", "Hunan", "Anhui", "Fujian", "Shandong", "Henan",
    "Hebei", "Shaanxi", "Liaoning", "Jilin", "Heilongjiang", "Tianjin",
    "Chongqing", "Jiangxi", "Guizhou", "Yunnan", "Shanxi", "Gansu",
    "Qinghai", "Hainan", "Inner Mongolia", "Guangxi", "Ningxia", "Xinjiang",
    "Tibet", "Hong Kong", "Macao"
]

# Â©öÂßªÁä∂ÂÜµÈÄâÈ°π
MARITAL_STATUS_OPTIONS = ['Single', 'Divorced-Single', 'Divorced-With-Child', 'Divorced-Without-Child', 'Widowed']

for i in range(1, 201):
    sid = f"58{str(i).zfill(6)}"
    college = random.choices(
        COLLEGES,
        weights=[30, 25, 20, 20, 5, 10],  # ÊØî‰æãÔºöÂïÜ30%„ÄÅ‰∫∫Êñá25%„ÄÅÁêÜ20%„ÄÅÂ∑•20%„ÄÅÂÖΩÂåª5%„ÄÅÂÖ∂‰ªñ10%
        k=1
    )[0]
    major = random.choice(MAJORS[college])
    
    # Ê†πÊçÆ‰∏ì‰∏öÂíåÈöèÊú∫Ê¶ÇÁéáÂÜ≥ÂÆöË∫´‰ªΩ
    if 'PhD' in major or random.random() < 0.05:  # 5%Ê¶ÇÁéáÂçöÂ£´Áîü
        identity = 'PhD'
        year = random.randint(1, 4)  # ÂçöÂ£´Áîü1-4Âπ¥
        # ÂçöÂ£´ÁîüÂπ¥ÈæÑÔºö25-35Â≤Å
        age = random.randint(25, 35)
    elif random.random() < 0.2:  # 20%Ê¶ÇÁéáÁ†îÁ©∂Áîü
        identity = 'Graduate'
        year = random.randint(1, 2)  # Á†îÁ©∂Áîü1-2Âπ¥
        # Á†îÁ©∂ÁîüÂπ¥ÈæÑÔºö22-28Â≤Å
        age = random.randint(22, 28)
    else:  # 75%Ê¶ÇÁéáÊú¨ÁßëÁîü
        identity = 'Undergraduate'
        year = random.randint(1, 4)  # Êú¨ÁßëÁîü1-4Âπ¥
        # Êú¨ÁßëÁîüÂπ¥ÈæÑÔºö18-24Â≤Å
        age = random.randint(18, 24)
    
    gender = random.choice(['M', 'F', 'X'])
    
    # ÁîüÊàêÂá∫ÁîüÊó•Êúü
    birth_year = datetime.now().year - age
    birth_month = random.randint(1, 12)
    birth_day = random.randint(1, 28)  # ÁÆÄÂåñÂ§ÑÁêÜÔºåÈÅøÂÖçÈó∞ÊúàÈóÆÈ¢ò
    birth_date = f"{birth_year}-{birth_month:02d}-{birth_day:02d}"
    
    # ÁîüÊàêË∫´È´òÔºàcmÔºâ- Ê†πÊçÆÊÄßÂà´Áï•Êúâ‰∏çÂêå
    if gender == 'M':
        height = round(random.uniform(160, 190), 2)
    elif gender == 'F':
        height = round(random.uniform(150, 180), 2)
    else:
        height = round(random.uniform(155, 185), 2)
    
    # ÁîüÊàê‰ΩìÈáçÔºàkgÔºâ
    weight = round(height * random.uniform(0.55, 0.75), 2)
    
    # ÈöèÊú∫ÈÄâÊã©Â©öÂßªÁä∂ÂÜµÔºà90%ÂçïË∫´Ôºâ
    marital_status = random.choices(
        MARITAL_STATUS_OPTIONS,
        weights=[90, 2, 1, 1, 1]  # 90%ÂçïË∫´ÔºåÂÖ∂‰ªñÂêÑÂç†‰∏ÄÂÆöÊØî‰æã
    )[0]
    
    # ÈöèÊú∫ÁîüÊàêÁêÜÊÉ≥‰º¥‰æ£ÊèèËø∞
    ideal_partners = [
        "Someone kind and caring with similar interests",
        "Ambitious person with great sense of humor",
        "Someone who loves traveling and outdoor activities",
        "Intellectual person who enjoys deep conversations",
        "Family-oriented person with strong values",
        "Adventurous spirit who loves trying new things"
    ]
    ideal_partner = random.choice(ideal_partners)
    
    # ÈöèÊú∫ÁîüÊàêÂÆ∂‰π°Ôºà‰∏≠ÂõΩÁúÅ‰ªΩÔºâ
    hometown = random.choice(PROVINCES)
    
    students.append({
        "student_id": sid,
        "name": f"Student_{i:03d}",
        "nickname": f"Nick_{i}",
        "gender": gender,
        "college": college,
        "year_of_study": year,
        "major": major,
        "email": f"{sid}@my.cityu.edu.hk",
        "wechat_id": f"wechat_{sid}",
        "bio": f"Hi! I'm from {college}. Love {random.choice(['movies', 'hiking', 'coding', 'reading', 'travel'])}.",
        "avatar_url": None,  # Â§¥ÂÉèËÆæ‰∏∫Á©∫
        "is_verified": 1,
        "is_active": 1,
        "birth_date": birth_date,
        "height": height,
        "weight": weight,
        "hometown": hometown,
        "marital_status": marital_status,
        "ideal_partner": ideal_partner,
        "identity": identity,
        "personal_photos": json.dumps([]),  # Á©∫Êï∞ÁªÑ
        "created_at": "2024-09-01 10:00:00",
        "updated_at": "2024-09-01 10:00:00"
    })

df_student = pd.DataFrame(students)
print("‚úÖ student.csv Â∑≤ÁîüÊàê (200Ë°å)")
df_student
```

```python
df_student.to_csv("student.csv", index=False)
```

```python
# student_interest
print("üéØ ÁîüÊàê student_interest.csv (ÊØè‰∏™Â≠¶Áîü5‰∏™Á±ªÂà´ÂÖ®Ë¶ÜÁõñ)...")

# ËØªÂèñÂ∑≤ÊúâÁöÑÊ†áÁ≠æÊï∞ÊçÆÔºàÂÅáËÆæ‰Ω†Â∑≤ÁªèÁîüÊàê‰∫Ü interest_tag.csvÔºâ
df_tags = pd.read_csv("interest_tag.csv")

# ÊåâÁ±ªÂà´ÂàÜÁªÑÊ†áÁ≠æ
category_tags = {}
for _, row in df_tags.iterrows():
    category = row['category']
    if category not in category_tags:
        category_tags[category] = []
    category_tags[category].append(row['tag_id'])

# ËØªÂèñÂ≠¶ÁîüÊï∞ÊçÆÔºàÂÅáËÆæ‰Ω†Â∑≤ÁªèÁîüÊàê‰∫Ü student.csvÔºâ
df_students = pd.read_csv("student.csv")
students = df_students['student_id'].tolist()

interest_data = []

for student_id in students:
    # ÊØè‰∏™Á±ªÂà´ÂàÜÈÖçÊ†áÁ≠æ
    for category, available_tags in category_tags.items():
        if category in ['MBTI', 'Zodiac']:
            # MBTI ÂíåÊòüÂ∫ßÊØè‰∫∫Âè™ËÉΩÊúâ‰∏Ä‰∏™
            selected_tag = random.choice(available_tags)
            interest_data.append({
                "student_id": student_id,
                "tag_id": selected_tag,
                "created_at": "2024-09-01 10:00:00"
            })
        else:
            # ÂÖ∂‰ªñÁ±ªÂà´ÊØè‰∫∫ÂèØ‰ª•Êúâ1-3‰∏™
            num_tags = random.randint(1, 3)
            selected_tags = random.sample(available_tags, min(num_tags, len(available_tags)))
            for tag_id in selected_tags:
                interest_data.append({
                    "student_id": student_id,
                    "tag_id": tag_id,
                    "created_at": "2024-09-01 10:00:00"
                })

df_interest = pd.DataFrame(interest_data)
df_interest
```
